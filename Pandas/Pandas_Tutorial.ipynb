Filtering - Using Conditionals to Filter Rows and Columns

```python
# a dictionary of lists
People = {
    'first': ['Karim', 'Mohamed', 'Adham', 'Ali'],
    'last': ['Adel', 'Ahmed', 'Mohsen', 'Adel'],
    'email': ['karimadel@gmail.com', 'mohamedahmed@gmail.com','adhammohsen@gmail.com', 'aliadel@gmail.com']
}

df = pd.DataFrame(People)
```

```python
filt = df['last'] == 'Adel'
filt
```

```python
df[filt]
```

```python
# it is better to use loc to filter rows
df.loc[filt]
```

```python
# the & operator 
filt = (df['last'] == 'Adel') & (df['first'] == 'Karim')
filt
```

```python
df.loc[filt]
```

```python
# the | operator 
filt = (df['last'] == 'Mohsen') | (df['first'] == 'Karim')
filt
```

```python
df.loc[filt]
```

```python
# the ~ operator is used to negate the filter
filt = ~((df['last'] == 'Mohsen') | (df['first'] == 'Karim'))
filt
```

```python
df.loc[filt]
```

Updating Rows and Columns - Modifying Data Within DataFrames

```python
# a dictionary of lists
People = {
    'first': ['Karim', 'Mohamed', 'Adham', 'Ali'],
    'last': ['Adel', 'Ahmed', 'Mohsen', 'Adel'],
    'email': ['karimadel@gmail.com', 'mohamedahmed@gmail.com','adhammohsen@gmail.com', 'aliadel@gmail.com']
}

df = pd.DataFrame(People)
```

```python
# Capitalizing the Column Names
df.columns = [x.capitalize() for x in df.columns]
df
```

```python
# Replacing the Underscores with Spaces but this is not recommended
df.columns = df.columns.str.replace('_', ' ')
df
```

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns = [x.lower() for x in df.columns]
df
```

```python
# Renaming a Single Column
df.rename(columns={'email': 'email_address'}, inplace=True)
df
```

```python
# Changing the Row Values
df.loc[2] = ['Adham', 'bebberbo', 'adhambebberbo@gmail.com']
df
```

```python
df.loc[filt, 'last_name'] = 'Hooda'
df
```

```python
# Setting the email address to uppercase
df['email_address'] = df['email_address'].str.upper()
df
```

Add/Remove Rows and Columns From DataFrames

```python
# a dictionary of lists
People = {
    'first': ['Karim', 'Mohamed', 'Adham', 'Ali'],
    'last': ['Adel', 'Ahmed', 'Mohsen', 'Adel'],
    'email': ['karimadel@gmail.com', 'mohamedahmed@gmail.com','adhammohsen@gmail.com', 'aliadel@gmail.com']
}

df = pd.DataFrame(People)
```

```python
# Adding a new column
df['full_name'] = df['first'] + ' ' + df['last']
df
```

```python
# Removing a column
df.drop(columns=['first', 'last'], inplace=True)
df
```

```python
# Splitting the full name into first and last name
df['full_name'].str.split(' ', expand=True)
```

```python
# Adding the first and last name to the dataframe
df[['first', 'last']] = df['full_name'].str.split(' ', expand=True)
df
```

```python
# Adding a single row
# The ignore_index is used to ignore the index of the new row
# append() method has been deprecated
new_col = pd.DataFrame({'first': ['Hassan']})
df = pd.concat([df, new_col], ignore_index=True)
df
```

```python
# Creating a new DataFrame 
People = {
    'first': ['Trent', 'Luis'],
    'last': ['Arnold', 'Diaz']
}
df2 = pd.DataFrame(People)
```

```python
df2["email"] = (df2["first"] + df2["last"] + "@gmail.com").str.lower()
df2
```

```python
# Appending the new DataFrame to the original DataFrame
df = pd.concat([df, df2], ignore_index=True, sort=True)
df
```

```python
# Removing a certain row
# df.drop(index = 4) # This is the first way to remove a row
filt = df['first'] == 'Hassan'
df.drop(index = df[filt].index, inplace= True) # This is the second way to remove a row
```

Cleaning Data - Casting Datatypes and Handling Missing Values

```python
import numpy as np
```

```python
people = {
    'first': ['Karim', 'Mohamed', 'Adham', 'Chris', np.nan, None, 'NA'], 
    'last': ['Adel', 'Ahmed', 'Mohsen', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['karimadel@gmail.com', 'mohamedahmed@email.com', 'adhammohsen@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['20', '19', '21', '36', None, None, 'Missing']
}
df = pd.DataFrame(people)
```

```python
df
```

```python
# Dropping the null values

df.dropna()
```

```python
df.dropna(axis='index', how='any') # Dafault arguments
# axis => index Drops the rows that have any missing values
# axis => column Drops the columns that have any missing values
```

```python
df.dropna(axis='index', how='all')
# all => Drops rows if it is all missing 
```

```python
# Dropping the rows that have the email set to None
df.dropna(axis='index', how='any', subset=['email'])
```

```python
# The lastname and email both are needed to be missing for the row to be dropped
df.dropna(axis='index', how='all', subset=['email', 'last'])
```

```python
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
df
```

```python
# Whether or not the values are classified as NaN
df.isna()
```

```python
df.fillna('MISSING') # Do this only for numerical Values NOT Strings
```

```python
df.dtypes  # Object is a string
```

```python
type(np.nan) # NAN is a float under the hood
```

```python
# When you change the TYPE of a Column To a numerical one , and it has missing values , Change it to FLOAT NOT INT 
df['age'] = df['age'].astype(float)
```

```python
df.dtypes
```

```python
df['age'].mean()
```
